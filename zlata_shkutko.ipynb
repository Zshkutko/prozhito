## Чего вы ждёте от неё, какие машина допустила ошибки, как их может показаться можным исправить. Почему современный NER допускает такие ошибки? Что вы вообще думаете об этом и смежном?
Жду наиболее приближенных к эталону результатов: как можно больше и правильнее относительно идеала. 
Примеры некоторых ошибок (в целом работает неплохо): 
1. "Maritain" определил как ORG, хотя это скорее PER, видимо из-за латиницы; странно, что в первом употреблении этого слова он его вообще никак не стал определять, видимо на это мог повлиять синтаксис (контекст). Исправить можно если добавить в модель некоторые знания о текстах на латинице (это видимо французский), или убрав эти слова, хотя так я думаю не стоит.
2. "Директора 2-го завода стройматериалов" определил как ORG, а это должность, видимо из-за "2-го завода...". Ну тут наверное надо усовершенствовать работу с контекстом и синтаксисом, может с морфологией еще где-то (?). Может захватил сюда директора из-за его регистра, хотя в отдельности от остального куска он его никак не помечает.
3. "Пол гниет и промокает" определил Пол как PER, это либо снова проблемы с регистром, либо там где-то встречалось иностарнное имя Пол и из-за этого такая ошибка. Его спасет только контекст. 
4. "Мороз" определил как PER - регистр?? может была такая фамилия...
5. "какой у Христа пессимизм" Христа чего-то не разметил :(

В итоге, наверное стоит покопаться немного в регистре, синтаксисе (чтобы фразы лишнее не захватывали), морфологии (падеж не везде учитывается видимо и поэто лишее берет), может есть какие-то штуки там работающие с контекстом типа n-грамм или что-то получше.
Допускает ошибки потому что сейчас, например, даже имя Пол встречается довольно часто (на кириллице), раньше, тем более в такое время, на вряд ли оно часто где-то фигурировало. Появилось очень много организаций с названиями на латинице, всяких компаний и тд, вообще латиница теперь очень часто в русских текстах встречается. Ну конечно появились всякие новые слова, сокращения, аббревеатуры и тд, это тоже повлияло. Язык сильно поменялся, да и стиль написания теперь везде разный, нет какого-то шаблона. 
Думаю, что есть всякие очевидные пути решения: большее количество текстов для обучения модели, их разнообразие (жанры, стили и все такое). Но можно еще в коде что-то усовершенствовать, чтобы программа анализировала более детально, больше внимания уделяла контексту, синтаксису, морфологии.
